```python
import datetime
import time
import csv
import tweepy
from datetime import datetime, timezone




client = tweepy.Client(bearer_token='AAAAAAAAAAAAAAAAAAAAAGRXHgEAAAAAn6gXAzo8mmvtiNglBCz52A2aNiU%3DHt7dYZVulpsHyHrY81ifbIKIrWo5EBsT3rFPyuqtlwTfSL7wRp')

csvData = [["created_at", "id", "text", "context_annotations", "geo"]]

cutoff_date = datetime(2021, 8, 23, tzinfo=timezone.utc)

while True:
    try:
        paginator = tweepy.Paginator(client.get_users_tweets, id='227073090', 
                                      tweet_fields=['context_annotations','created_at','geo'], 
                                      max_results=100)
        for tweet in paginator.flatten(limit=3500):
            tweet_date = tweet.created_at
            if tweet_date < cutoff_date:
                break
            csvData.append([tweet.created_at, tweet.id, tweet.text, 
                            tweet.context_annotations if tweet.context_annotations else None, 
                            tweet.geo if tweet.geo else None])
        break  # This will break the while loop after successfully fetching the tweets

    except tweepy.TooManyRequests:
        print("Rate limit reached. Sleeping for 15 minutes.")
        time.sleep(15*60)  # Sleep for 15 minutes

# Write the CSV data
with open('mgptweets2.csv', 'w', newline='', encoding='utf-8') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerows(csvData)

csvFile.close()
```
